```python
reports = dict()
for size in [300,400,500,600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400]:
    print(size)
    inData = analysis.load_results(PATH ='transit_results', EXP_NAME = 'size_{}'.format(size))
    inData = analysis.make_report(inData)
    inData.transitize.report['size']=size
    reports[size] = inData.transitize.report.copy()
    
```

```python
fig, ax = plt.subplots(1,4,figsize = (16,4))

to_plot= dict()
for k,v in reports.items():
    to_plot[k]=v.T.occupancy
df =pd.DataFrame(to_plot)

df = df.T[[0,1,2,3]]
df.columns = ['p','d2d','s2s','ms']
df.index = df.index*6
df.plot(ax = ax[0])
ax[0].set_ylabel('occupancy')
ax[0].set_xlabel('demand')
#ax[0].set_title('Occupancy of respective pooling levels growing with the demand')


to_plot= dict()
for k,v in reports.items():
    to_plot[k]=v.T.u_pax/v.T.u_pax[0]
df =pd.DataFrame(to_plot)

df = df.T[[0,1,2,3]]
df.columns = ['private','d2d','s2s','ms']
df.index = df.index*6
df.plot(ax = ax[1])
ax[1].set_ylabel('relative utility improve')
ax[1].set_xlabel('demand')

to_plot= dict()
for k,v in reports.items():
    to_plot[k]=v.T.u_veh/v.T.u_veh[0]
df =pd.DataFrame(to_plot)

df = df.T[[0,1,2,3]]
df.columns = ['private','d2d','s2s','ms']
df.index = df.index*6
df.plot(ax = ax[2])
ax[2].set_ylabel('relative vehicle hours savings')
ax[2].set_xlabel('demand')

to_plot= dict()
for k,v in reports.items():
    to_plot[k]=v.T.fare/v.T.fare[0]
df =pd.DataFrame(to_plot)

df = df.T[[0,1,2,3]]
df.columns = ['private','d2d','s2s','ms']
df.index = df.index*6
df.plot(ax = ax[3])
ax[3].set_ylabel('relative fare reduction')
ax[3].set_xlabel('demand')


fig.tight_layout()
```
