```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
plt.style.use('fast')
plt.rcParams["font.family"] = "Helvetica"
colors = sns.color_palette("muted")
plt.rcParams['figure.figsize'] = [6, 6]
plt.rcParams["mathtext.fontset"] = "cm"
```

```python
fig, ax = plt.subplots()
ax.plot([0,0.9],[0.1,0.9], alpha = 0)
#scheduled
ax.arrow(0.2, 0.1, 0.6, 0.5, lw = 2, color = 'black', ls = '-', head_width=0.02, head_length=0.02, head_starts_at_zero=False, shape='full',
                      length_includes_head=True)
#wait delay
plt.plot([0.2, 0.2], [0.1, 0.2], lw = 1, color = 'black', ls = '--')
plt.plot([0.2, 0.8], [0.2, 0.7], lw = 1, color = 'black', ls = '--')
plt.plot([0.8, 0.8], [0.7, 0.9], lw = 1, color = 'black', ls = '--')

#scheduled delay

#in vehicle scheduled delay
ax.arrow(0.2, 0.2, 0.6, 0.7, lw = 2, color = 'black', ls = '-', head_width=0.02, head_length=0.02, head_starts_at_zero=False, shape='full',
                      length_includes_head=True)


ax.text(0.48,0.36,'scheduled ride', rotation=42)
ax.text(0.48,0.56,'actual ride', rotation=53)
ax.text(0.02,0.15,'origin delay')
ax.text(0.02,0.55,'')
ax.text(0.02,0.8,'in vehicle \ndelay')
ax.text(0.02,0.4,'scheduled  \nride time')
ax.text(0.06,0.11,r'$WO_i$')
ax.text(0.06,0.76,r'$WV_i$')
ax.text(0.06,0.36,r'$\bar{T}_i$')


ax.set_xlabel('space')
ax.set_ylabel('time')
ax.set_xticks([0.2,0.8])
ax.set_xticklabels(['pickup','drop-off'])
ax.set_yticks([0.1,0.2,0.6,0.7,0.9])
ax.set_yticklabels([r'$\bar{P}_i $',r'$O_i$',r'$\bar{D}_i $','',r'$D_i$'])


ax.grid()
plt.tight_layout()
plt.savefig('fig2.png')
```
